```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

loading the dataset

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

printing the first 5 rows of the dataset

```python
credit_card_data.head()
```

printing the last 5 rows of the dataset

```python
credit_card_data.tail()
```

lets have a look on some information of the dataset

```python
credit_card_data.info()
```

lets have a look on the number of missing values in each column

```python
credit_card_data.isnull().sum()
```

lets have a look on legit transactions (having the class value == 0 then legit and if it is == 1 then fraud transaction)

```python
credit_card_data['Class'].value_counts()
```

storing the legit and fraud values in different variable


```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

printing the shape of the legit and fraud

```python
print(legit.shape)
print(fraud.shape)
```

some statistical measures of the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

compare the values for both transactions


```python
credit_card_data.groupby('Class').mean()
```

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=169)
```

now concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

lets have a look on the new_dataset

```python
print(new_dataset)
```

splitting the data into Features and Target

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

lets split out new_dataset into training and testing

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

lets have a look on the shape of the training and testing dataset after splitting

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training
Logistic Regression because binary classification problem that either the transaction is legit or not

```python
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(X_train, Y_train)
```

lets evaluate the model on the basis of accuracy (accuracy on training data)

```python
X_train_prediction = model.predict(X_train)
```

lets compare X_train_prediction and Y_train

```python
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

lets have a look on accuracy score

```python
print('Accuracy of the training model is : ', training_data_accuracy)
```

lets check the accuracy of the test dataset

```python
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy of the testing model is : ', test_data_accuracy)
```
